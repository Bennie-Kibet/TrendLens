# Exploratory Data Analysis (EDA) - English-Dholuo Corpus

This notebook explores the MT560 dataset for English-Dholuo translation.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/mt560_clean.csv')
print(f"Total rows: {len(df)}")
df.head()
```

```python
# Sentence length distribution
df['en_len'] = df['english'].apply(lambda x: len(str(x).split()))
df['luo_len'] = df['dholuo'].apply(lambda x: len(str(x).split()))

plt.figure(figsize=(12, 6))
sns.histplot(df['en_len'], label='English', kde=True, color='blue')
sns.histplot(df['luo_len'], label='Dholuo', kde=True, color='green')
plt.title('Sentence Length Distribution')
plt.legend()
plt.show()
```
